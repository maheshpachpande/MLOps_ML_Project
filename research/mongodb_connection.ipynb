```python
DB_NAME = 'us_visa'
COLLECTION_NAME = 'visa_data'
```

```python
from dotenv import load_dotenv
import os

load_dotenv()
    
mongo = os.getenv("URL")
```

```python
import pymongo
import certifi

ca = certifi.where()

client = pymongo.MongoClient(mongo, tlsCAFile=ca)
db = client[DB_NAME]
collection = db[COLLECTION_NAME]
```

```python
import pandas as pd

df = pd.DataFrame(list(collection.find({})))
```

```python
if "_id" in df.columns.to_list():
    df = df.drop(columns=["_id"])
```
